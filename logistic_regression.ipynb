```python
dummies = pd.get_dummies(df[['salary','Department']]).astype(int)
dummies.head()
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged.head()
```

```python
final = merged.drop(['Department','salary','Department_product_mng','salary_high'],axis='columns')
final
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='satisfaction_level',y='left',data=df,palette='Blues_d')
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='number_project',y='left',data=df,palette='Blues_d')
```

```python
X = final[['average_montly_hours','satisfaction_level','salary_low','salary_medium']]
X
```

```python
y = final.left
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.16)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
